<font size="5"><b>Fill dummy values</b></font><br/>

Analyzing the missingness of a variable against another variable helps you determine any relationships between missing and non-missing values. This is very similar to how you found correlations of missingness between two columns in the previous exercises.

In this exercise, you're going to write a function which fills in dummy values in the DataFrame and in the next exercise you are going to plot a scatter plot to analyze if there is any correlation of missingness against another column. The operations involve shifting and scaling the column range with a scaling factor.

The function <mark>rand()</mark> has been imported for you from <mark>numpy.random</mark>.

```python
# Calculate the number of missing values in each column of the dummy DataFrame.

def fill_dummy_values(df):
  df_dummy = df.copy(deep=True)
  for col_name in df_dummy:
    col = df_dummy[col_name]
    col_null = col.isnull()    
    # Calculate number of missing values in column 
    num_nulls = col_null.sum()
  return df_dummy
```

```python
# Calculate the range of the column.

def fill_dummy_values(df):
  df_dummy = df.copy(deep=True)
  for col_name in df_dummy:
    col = df_dummy[col_name]
    col_null = col.isnull()    
    # Calculate number of missing values in column 
    num_nulls = col_null.sum()
    # Calculate column range
    col_range = col.max() - col.min()
  return df_dummy
```

```python
# import rand from numpy.random
from numpy.random import rand

# Calculate random values with the size of num_nulls.

def fill_dummy_values(df):
  df_dummy = df.copy(deep=True)
  for col_name in df_dummy:
    col = df_dummy[col_name]
    col_null = col.isnull()    
    # Calculate number of missing values in column 
    num_nulls = col_null.sum()
    # Calculate column range
    col_range = col.max() - col.min()
    # Calculate random values of size num_nulls
    dummy_values = (rand(num_nulls) - 2)
  return df_dummy
```

- In your function definition, set the default value of scaling_factor to be 0.075
- Next, scale your dummy values by scaling_factor and multiply them by col_range. The minimum col.min() has been already added for you.

```python
def fill_dummy_values(df, scaling_factor=0.075):
  df_dummy = df.copy(deep=True)
  for col_name in df_dummy:
    col = df_dummy[col_name]
    col_null = col.isnull()    
    # Calculate number of missing values in column 
    num_nulls = col_null.sum()
    # Calculate column range
    col_range = col.max() - col.min()
    # Scale the random values to scaling_factor times col_range
    dummy_values = (rand(num_nulls) - 2) * scaling_factor * col_range + col.min()
    col[col_null] = dummy_values
  return df_dummy
```

<b>Congratulations! You just built a function that automates creating dummy values for missing data! In the next exercise, you'll use this function to visualize the relationship between missing values in a DataFrame.</b>

<font size="5"><b>Generate scatter plot with missingness</b></font><br/>

In this exercise you'll create a scatter plot consisting of both missing and non-missing values. You will utilize the function <mark>fill_dummy_values()</mark> which you created in the previous exercise for filling in dummy values in the DataFrame <mark>diabetes_dummy</mark>.

For setting different colors to the missing and non-missing values, you can simply add the nullity, or the sum of null values of both respective columns that you are plotting calculated using the <mark>.isnull()</mark> method. The nullity returns a Series of <mark>True</mark> or <mark>False</mark> where:

- <mark>True</mark> → Either <mark>col1</mark> or <mark>col2</mark> or both values are missing.
- <mark>False</mark> → Neither of <mark>col1</mark> and <mark>col2</mark> values are missing.
The DataFrame diabetes and the function <mark>fill_dummy_values()</mark> have been loaded for your usage.

<font size="2"><b>Instructions</b></font><br/>

- Fill dummy values in <mark>diabetes_dummy</mark> using <mark>fill_dummy_values()</mark>.
- Sum the nullity of <mark>'BMI'</mark> and <mark>'Skin_Fold'</mark>.
- Create a scatter plot of <mark>'Skin_Fold'</mark> vs. <mark>'BMI'</mark>.

```python
# Fill dummy values in diabetes_dummy
diabetes_dummy = fill_dummy_values(diabetes)

# Sum the nullity of Skin_Fold and BMI
nullity = diabetes.Skin_Fold.isnull() + diabetes.BMI.isnull()

# Create a scatter plot of Skin Fold and BMI 
diabetes_dummy.plot(x='Skin_Fold', y='BMI', kind='scatter', alpha=0.5, 
                    
                    # Set color to nullity of BMI and Skin_Fold
                    c=nullity, 
                    cmap='rainbow')

plt.show()
```

<b>Awesome work! Make sure you take a deep look at how the missing values of Skin_Fold and BMI interact.</b>

In this exercise, you'll listwise delete the rows where the <mark>Glucose</mark> column has missing values. The <mark>diabetes</mark> DataFrame and the <mark>missingno</mark> package as <mark>msno</mark> has already been loaded for you.
